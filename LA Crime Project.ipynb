# Loading Data

```python
import pandas as pd
import matplotlib.pyplot as plt
crime = pd.read_csv('C:/Users/playe/Crime_Data_from_2020_to_Present_20241020.csv')
```

```python
# Checking Nulls
crime.isna().sum()
```

```python
#Dropping non-relavant data
crime_df = crime.drop(labels=['Mocodes', 'Weapon Used Cd', 'Weapon Desc', 
                              'Crm Cd 2', 'Crm Cd 3', 'Crm Cd 4', 'Cross Street'], axis=1)
crime_df
```

```python
# Sorting by date reported
crime_df.sort_values(by='Date Rptd')
```

```python
# Formatting date reported and date occured
crime_df['Date Rptd'] = pd.to_datetime(crime_df['Date Rptd'],format= 'mixed', dayfirst=True)
crime_df['DATE OCC'] = pd.to_datetime(crime_df['Date Rptd'],format= 'mixed', dayfirst=True)
crime_df.info()
```

```python
# Creating month and year specific columns
crime_df['year'] = crime_df['DATE OCC'].dt.year
crime_df['month'] = crime_df['DATE OCC'].dt.month
crime_df.head()
```

```python
# Sorting by date reported
crime_df.sort_values(by='Date Rptd')
```

### Analyzing Data

```python
# bar chart of crime totals over the year
crime_df['year'].value_counts().sort_index().plot(kind = 'bar')
```

```python
# Removing incomplete data for the month & year of 12/2024
df = crime_df[~((crime_df['month'] == 12) & (crime_df['year'] == 2024))]

# Sorting by date occured
df.sort_values(by = 'DATE OCC')

### Plot of crimes over time by year & month
crime_over_time = df.groupby(['year', 'month']).size().plot()
```

```python
# Finding counts of victim ages
crime_df['Vict Age'].value_counts()

# Removing ages less than 0
df_age = crime_df[crime_df['Vict Age'] > 0]
df_age['Vict Age'].value_counts()
```

```python
# MAtplotlib histogram of crime numbers per victim age
plt.hist(df_age['Vict Age'], edgecolor='white', bins=15)
plt.show()
```

```python
crime_count_per_day = df.groupby(by = [df['DATE OCC'], 'Crm Cd Desc']).size()

top_20_crimes = crime_count_per_day.groupby('Crm Cd Desc').mean().reset_index(name='avg_daily_count')\
    .sort_values(by= 'avg_daily_count', ascending=False).head(20)

top_20_crimes_sorted = top_20_crimes.sort_values(by='avg_daily_count', ascending=True)

plt.barh(top_20_crimes_sorted['Crm Cd Desc'], top_20_crimes_sorted['avg_daily_count'])
```

```python
# area with highest number of crimes during nighttime 8pm
nighttime_crime = df[(df['TIME OCC'] >= 2000) | (df['TIME OCC'] < 300)]
area_crime_count = nighttime_crime.groupby(by = nighttime_crime['AREA NAME']).size()

top_10_areas = area_crime_count.groupby('AREA NAME').mean().reset_index(name='avg_daily_count')\
    .sort_values(by= 'avg_daily_count', ascending=False).head(10)

top_10_areas_sorted = top_10_areas.sort_values(by='avg_daily_count', ascending=True)

plt.barh(top_10_areas_sorted['AREA NAME'], top_10_areas_sorted['avg_daily_count'])
```

```python
## Combining charts into matplotlib subplot
plt.style.available
plt.style.use('ggplot')
```

```python
plt.figure(figsize= (14,8))

#Plot1 = Line Chart
plt.subplot(2, 2, 1)
crime_plt = df.groupby(['year', 'month']).size().plot(color = 'green', linewidth = 2)
crime_plt.set_xlabel('Year - Month', fontsize = 10)
crime_plt.set_title('Crime Over Time')
plt.tight_layout()

#Plot2 = Histogram
plt.subplot(2, 2, 2)
plt.hist(df_age['Vict Age'], edgecolor='white', bins=15)
plt.xlabel('Age', fontsize = 10)
plt.title('Distribution of Victim Age')
plt.tight_layout()

#Plot3 = Horizontal Bar
plt.figure(figsize=(14,6))
plt.barh(top_10_areas_sorted['AREA NAME'], top_10_areas_sorted['avg_daily_count'], color='steelblue')
plt.xlabel('Crime Count', fontsize = 10)
plt.title('Top 10 Crime Areas During Nighttime')
plt.tight_layout()

#Plot4 = Horizontal Bar
plt.figure(figsize=(14,6))
plt.barh(top_20_crimes_sorted['Crm Cd Desc'], top_20_crimes_sorted['avg_daily_count'], color='orange')
plt.xlabel('Crime Count', fontsize = 10)
plt.title('Top 20 Crimes')
plt.tight_layout()
plt.show()
```
